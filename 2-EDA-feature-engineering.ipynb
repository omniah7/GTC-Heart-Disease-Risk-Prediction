# FEATURE ENGINEERING

```python
# Age × Resting Blood Pressure
data["Age_BP_Interaction"] = data["age"] * data["trestbps"]

# Age × Cholesterol
data["Age_Chol_Interaction"] = data["age"] * data["chol"]

# Cholesterol × Resting Blood Pressure
data["Chol_BP_Interaction"] = data["chol"] * data["trestbps"]
data["BP_Chol_Ratio"] = data["trestbps"] / data["chol"]
```

```python
data["Age_Squared"] = data["age"] ** 2
data["Chol_Squared"] = data["chol"] ** 2

data.to_csv("data/engineered_data.csv", index=False)

data.head()
```
