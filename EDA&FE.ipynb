FEATURE ENGINEERING

```python
# Age  Resting Blood Pressure
data["Age_BP_Interaction"] = data["age"] * data["trestbps"]
# Age  Cholesterol
data["Age_Chol_Interaction"] = data["age"] * data["chol"]
# Cholesterol  Resting Blood Pressure
data["Chol_BP_Interaction"] = data["chol"] * data["trestbps"]
```
